# Practicing Pandas

## 2.8

In this we'll be investigating the carbon footprint of different foods. 

We'll be leveraging a dataset compiled by [Kasia Kulma](https://r-tastic.co.uk/post/from-messy-to-tidy/) and contributed to [R's Tidy Tuesday project](https://github.com/rfordatascience/tidytuesday).

Start by importing pandas with the alias `pd`.

````python
import pandas as pd
````

## 2.9 

The dataset we'll be working with has the following columns:

|column      |description |
|:-------------|:-----------|
|country       | Country Name |
|food_category | Food Category |
|consumption   | Consumption (kg/person/year) |
|co2_emmission | Co2 Emission (Kg CO2/person/year) |


Import the dataset as a dataframe named `df` from this url: <https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv>

````python
url="https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-18/food_consumption.csv"
df=pd.read_csv(url)
print(df.head())
````

## 2.10 

How many rows and columns are there in the dataframe?

````python
rows, columns=df.shape
print("# of rows:",rows)
print("# of columns:",columns)
````

## 2.11

What is the mean `co2_emission` of the whole dataset?

````python
co2_avg=df['co2_emmission'].mean()
print("Mean CO2 Emission of whole dataset:",co2_avg)
````

## 2.12

What is the maximum `co2_emmission` in the dataset and which food type and country does it belong to?

````python
max_co2_emmission = df['co2_emmission'].max()
query=df[df['co2_emmission']==max_co2_emmission]
country1= query['country'].values[0]
food_type= query['food_category'].values[0]
print(food_type,"in",country1, "has maximum CO2 emmissions of", max_co2_emmission)
````

## 2.13

How many countries produce more than 1000 Kg CO2/person/year for at least one food type?

````python
k_emmissions=df[df['co2_emmission']>1000]
num_countries_with_emmissions=k_emmissions['country'].nunique()
countries_with_emmissions=k_emmissions['country'].unique()
print("There are ",num_countries_with_emmissions, "countries that produce more than 1000kg CO2/person/year for at least one food type.")
print(countries_with_emissions)
````

## 2.14

Which country consumes the least amount of beef per person per year?


````python
df_beef=df[df['food_category']=='Beef']
min_beef=df_beef['consumption'].idxmin()
country_min_beef=df_beef.loc[min_beef, 'country']
min_consumption=df_beef.loc[min_beef, 'consumption']
print("The country that consumes the least amount of beef per year is", country_min_beef, ", consuming", min_consumption, "of beef per person per year.")
````

## 2.15

What is the total emissions of all the meat products (Pork, Poultry, Fish, Lamb & Goat, Beef) in the dataset combined?

````python
meat_products = ["Pork","Poultry","Fish","Lamb & Goat", "Beef"]
df_meat=df[df['food_category'].isin(meat_products)]
total_emmissions_meat=df_meat['co2_emmission'].sum()
print("The total emissions of all the meat products in the dataset combined is", total_emmissions_meat)
````

# 2.16 

What is the total emissions of all other (non-meat) products in the dataset combined?

````python
meat_products = ["Pork","Poultry","Fish","Lamb & Goat", "Beef"]
df_notmeat=df[~df['food_category'].isin(meat_products)]
total_emmissions_notmeat=df_notmeat['co2_emmission'].sum()
print("The total emissions of all the non-meat products in the dataset combined is", total_emmissions_notmeat)
````
